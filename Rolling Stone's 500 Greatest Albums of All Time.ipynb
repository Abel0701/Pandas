```python
import pandas as pd
```

```python
# Cargamos el archivo csv
Rolling_Stone = pd.read_csv('albumlist.csv', encoding = 'ISO-8859-1')
Rolling_Stone
```

```python
# Mostramos los álbumes que se lanzaron en un año específico
def anno_especifico(df, n):
    resultado = df[df['Year'] >= n].sort_values('Year')
    return resultado
```

```python
anno_especifico(Rolling_Stone, 1965)
```

```python
# Filtramos por genero en particular
def genero(df, gen):
    resultado = df[df['Genre'] == gen]
    return resultado
```

```python
genero(Rolling_Stone, 'Rock')
```

```python
# Mostramos los diez primeros álbumes
Rolling_Stone.iloc[:10]
```

```python
# Buscamos álbum por artista específico
def artista(df, art):
    resultado = df[df['Artist'] == art]
    return resultado
```

```python
artista(Rolling_Stone, 'The Beach Boys')
```

```python
# Cantidad de géneros unicos
Rolling_Stone['Genre'].nunique()
```

```python
# Contamos los álbumes por década
def decada_album(df, y1):
    if df[(df['Year'] >= y1) & (df['Year'] < y1 + 10)].empty:
        print("No hay resultados")
    else:
        resultado = df[(df['Year'] >= y1) & (df['Year'] < y1 + 10)].sort_values('Year')
        print(f"Álbumes de la decada de {y1}")
        return resultado
```

```python
decada_album(Rolling_Stone, 1970)
```

```python
# Estadísticas por género más común
genero_cantidad = Rolling_Stone['Genre'].value_counts()
genero_mas_comun = genero_cantidad.idxmax()
cantidad_albumes = genero_cantidad.max()
total_albumes = genero_cantidad.sum()

print("Género más común:", genero_mas_comun)
print("Cantidad de álbumes en el género más común:", cantidad_albumes)
print("Total álbumes en la lista:", total_albumes)
```
